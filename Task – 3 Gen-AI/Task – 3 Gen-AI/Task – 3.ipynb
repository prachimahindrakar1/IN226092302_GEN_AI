```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
```

```python
output = generator("Explain resume screening in simple terms", max_length=50)

print(output[0]["generated_text"])
```

```python
job_description = """
Looking for a Data Scientist.

Skills required:
- Python
- Machine Learning
- Deep Learning
- SQL
- Pandas, NumPy

Experience: Minimum 2 years
"""
```

```python
# Strong Resume
resume_strong = """
Name: Prachi
Experience: 3 years
Skills: Python, Machine Learning, Deep Learning, SQL, Pandas, NumPy
"""

# Average Resume
resume_avg = """
Name: Riya
Experience: 1 year
Skills: Python, Pandas, SQL
"""

# Weak Resume
resume_weak = """
Name: Amit
Experience: Fresher
Skills: MS Excel, Basic Computer Knowledge
"""
```

```python
print(job_description)
print(resume_strong)
```

```python
def extract_info(resume):

    prompt = f"""
    Extract the following from this resume:

    1. Skills
    2. Experience

    Resume:
    {resume}

    Answer clearly.
    """

    output = generator(prompt, max_length=100)

    return output[0]["generated_text"]
```

```python
print(extract_info(resume_strong))
```

```python
print(extract_info(resume_avg))
print(extract_info(resume_weak))
```
